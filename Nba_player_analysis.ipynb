```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dados fornecidos
data = {
    "name": ["LeBron James", "Stephen Curry", "Giannis Antetokounmpo", "Kevin Durant", "Joel Embiid"],
    "team": ["Los Angeles Lakers", "Golden State Warriors", "Milwaukee Bucks", "Phoenix Suns", "Philadelphia 76ers"],
    "position": ["Small Forward", "Point Guard", "Power Forward", "Small Forward", "Center"],
    "points": [25000, 23000, 20000, 22000, 18000],
    "xp": [1500, 1400, 1200, 1300, 1100],
    "rebounds": [7000, 5000, 8000, 6000, 7500],
    "assists": [7000, 6000, 3000, 4000, 2000],
    "fgm": [9000, 8000, 7000, 8500, 6000],
    "fga": [15000, 13000, 11000, 14000, 10000],
    "ftm": [5000, 4000, 3000, 3500, 2000],
    "fta": [7000, 5000, 4000, 5000, 3000],
    "3pm": [2000, 3000, 1000, 1500, 500],
    "3pa": [4000, 5000, 2000, 3000, 1000],
    "steals": [2000, 1500, 1200, 1000, 800],
    "blocks": [1000, 500, 1500, 800, 2000],
    "games_played": [1300, 1200, 1000, 1100, 900]
}

# Convertendo para DataFrame
df = pd.DataFrame(data)

# Definindo o estilo do grfico
sns.set(style="whitegrid")

# Grfico de barras para Pontos, Assistncias e Rebotes por Posio
fig, axs = plt.subplots(1, 3, figsize=(20, 15))

# Grfico de Pontos
sns.barplot(x='position', y='points', data=df, ax=axs[0], palette='Blues_d')
axs[0].set_title('Pontos por Posio')
axs[0].set_xlabel('Posio')
axs[0].set_ylabel('Pontos')

# Grfico de Assistncias
sns.barplot(x='position', y='assists', data=df, ax=axs[1], palette='Greens_d')
axs[1].set_title('Assistncias por Posio')
axs[1].set_xlabel('Posio')
axs[1].set_ylabel('Assistncias')

# Grfico de Rebotes
sns.barplot(x='position', y='rebounds', data=df, ax=axs[2], palette='Reds_d')
axs[2].set_title('Rebotes por Posio')
axs[2].set_xlabel('Posio')
axs[2].set_ylabel('Rebotes')

# Ajustando a disposio dos grficos
plt.tight_layout()
plt.show()
```

Na pontuao, ala amador, se sai melhor que as outras posies. Em assistncia point guard se sai melhor, e em rebotes point forward se sai melhor. Apesar que em pontuao, point guard tambm possui um bom score, assim como small foward se sai bem em assistncia e piv se sai bem em rebotes.
